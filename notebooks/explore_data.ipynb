```python
import gzip
import csv
import itertools
import pprint

DATA_PATH = '../data/imdb-dataset.csv.gz'

def yield_all_data_lines():
    with gzip.open(DATA_PATH, mode='rt', encoding='utf-8') as file:
        reader = csv.reader(line for line in file.readlines())
        return (row for row in reader)
    
all_data = yield_all_data_lines()
pprint.pprint(list(itertools.islice(all_data, 10)))
```
